#data preprocessing

```python
df1.info()
#onr column is a categorical column need to convert it into numerical
```

```python
#aplied label encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df1['ocean_proximity']=le.fit_transform(df1['ocean_proximity'])
```

```python
df1['ocean_proximity']
```

```python
#data normalisation
#applying minmaxscaler
for col in df1.columns:
  df1[col]=(df1[col]-df1[col].min())/(df1[col].max()-df1[col].min())
```

#Feature selection

```python
df1.columns
```

```python
X=df1[['longitude', 'latitude', 'housing_median_age', 'total_rooms',
       'total_bedrooms', 'population', 'households', 'median_income',
       'ocean_proximity']]
y=df1['median_house_value']
```

```python
#test train splitting
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
ypre=pd.DataFrame(y_pred)
ypre.head(10)
```

```python
#y_test.head(10)
```

```python
from sklearn.metrics import mean_squared_error
mae=mean_squared_error(y_test,y_pred)
mae
```

```python
from sklearn import metrics
r2_lr = metrics.r2_score(y_test,y_pred)
r2_lr
```
